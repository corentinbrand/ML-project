# Classification Heart_Disease_Risk

- **Titre de slide conseillé :** Classification : modèles comparés et résultats
- **Contenu à afficher :**
  - Modèles testés :
    - régression logistique simple
    - logistique pénalisée LASSO/Ridge/Elastic Net
    - CART
    - bagging / random forest
    - boosting / AdaBoost
    - SVM
    - réseau de neurones MLP
  - Résultats représentatifs :
    - régression logistique simple : accuracy ≈ 0,732
    - LASSO : accuracy ≈ 0,732, sélection faible
    - CART : accuracy test ≈ 0,703
    - CART élagué : ≈ 0,708
    - random forest : ≈ 0,719 à 0,727 selon notebook/split
    - AdaBoost : ≈ 0,730
    - SVM : ≈ 0,729, ROC-AUC ≈ 0,79 dans le notebook SVM
    - MLP : erreur ≈ 0,273, donc accuracy ≈ 0,727
- **Ce qu’on raconte réellement :**
  - “Le premier résultat important est que les modèles complexes ne font pas exploser la performance. Les meilleurs modèles tournent autour de 72-73 % d’accuracy.”
  - “Cela suggère que le signal de classification est réel mais limité : le risque cardiaque synthétique dépend d’une combinaison de facteurs, mais il reste bruité ou partiellement recouvrant.”
- **Messages importants :**
  - Le meilleur modèle n’est pas écrasant.
  - La régression logistique est une baseline très forte.
  - Le boosting et le SVM sont compétitifs mais avec moins d’interprétabilité.
- **Points techniques à valoriser :**
  - SVM : choix du noyau via CV ; linéaire ou RBF selon notebook, performances proches.
  - CART : interprétable mais variance élevée.
  - Random forest : réduction de variance par agrégation.
  - Boosting : correction séquentielle des erreurs, bon compromis performance.
  - MLP : pas de gain majeur, convergence parfois imparfaite.
- **Pièges/questions possibles du jury :**
  - Pourquoi un modèle plus complexe ne gagne pas beaucoup ? Parce que la frontière de décision n’est pas nécessairement très non linéaire, et le signal disponible plafonne.
  - Pourquoi garder la logistique ? Pour son interprétabilité et sa performance quasi optimale.
- **Transition orale :**

# Questions Probables Du Jury

- **Pourquoi séparer train/test ?**
  - Pour estimer la généralisation sur des données non vues.
- **Pourquoi validation croisée ?**
  - Pour choisir les hyperparamètres sans utiliser le test.
- **Pourquoi standardiser ?**
  - Parce que SVM, kNN, modèles pénalisés et réseaux sont sensibles à l’échelle.
- **Pourquoi l’accuracy n’est pas suffisante ?**
  - La cible est un peu déséquilibrée ; F1, balanced accuracy et ROC-AUC donnent une vision plus complète.
- **Pourquoi les modèles complexes ne gagnent pas beaucoup ?**
  - Le signal disponible est limité ou déjà capturé par des relations assez simples.
- **Pourquoi la régression LDL est meilleure avec des modèles linéaires ?**
  - À cause de la forte relation entre `Cholesterol_Total` et `Cholesterol_LDL`.
- **Peut-on interpréter les importances comme causalité ?**
  - Non, ce sont des associations prédictives.
- **Pourquoi ne pas supprimer les outliers ?**
  - Ils restent plausibles médicalement et ne sont pas des erreurs manifestes.
- **Pourquoi ne pas retenir l’ACP ?**
  - Elle ne sépare pas bien les classes et disperse l’information sur plusieurs composantes.
- **Quel modèle recommander ?**
  - Classification : logistique/LASSO pour interprétabilité, boosting/SVM pour performance légèrement supérieure.
  - Régression : Lasso ou régression linéaire sélectionnée.
